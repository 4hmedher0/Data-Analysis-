```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_excel('COVID-19-geographic-disbtribution-worldwide-2020-12-14.xlsx')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.rename(columns={'Cumulative_number_for_14_days_of_COVID-19_cases_per_100000':'14daysper1000','countriesAndTerritories':'country','continentExp':'continent'},inplace=True)
```

```python
df1=df[df.month!=12]
```

```python
df1
```

```python
df.query('month==12 and year==2020').head(5)
```

```python
df1.drop(['geoId'],axis=1,inplace=True)
```

```python
df1.rename(columns={'countryterritoryCode':'countryCode'},inplace=True)
```

```python
df1group=df1.groupby('country')['cases','deaths'].sum()
```

```python
df1group['moralityRate']=df1group['cases']/df1group['deaths']
```

```python
df1group
```

```python
df1.isna().sum().sum()/len(df1)
```

### Since the percentage of null values is less than 1%, we drop the 
### null values in the following step

```python
df1.dropna(inplace=True)
```

```python
df1.isna().sum().sum()/len(df1)
```

```python
ax=df1group['cases'].sort_values(ascending=False).head(20).plot(kind='bar',figsize=(20,10))
ax.set_title('Bar chart depicting countries with the highest mortality rates for COVID-19')
```

```python
ax=df1group['cases'].sort_values(ascending=False).head(10).plot(kind='pie',autopct='%.2f%%',figsize=(20,10))
ax.set_title(' Pie chart depicting the share of the top ten countries by COVID-19 cases')
```

```python
df1group['moralityRate'].sort_values(ascending=True).head(5)
```

```python
ax=(df1group['deaths']/1000).sort_values(ascending=False).head(10).plot(kind='bar',figsize=(15,10),color='red',alpha=0.7)
for p in ax.patches:
 ax.annotate(p.get_height(),(p.get_x()+p.get_width()/2,p.get_height()),ha='center',va='bottom')

ax.set_xlabel('Country',size=16)
ax.set_ylabel('Number of Deaths by K',size=16)
ax.set_title('Countries suffering the most fatalities from COVID-19 BY Thonands ',size=17,color='b')
```

```python
df2group=df1.groupby('month')['cases','deaths'].sum()
```

```python
df2group
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(18,8))
sns.set()
ax1=sns.lineplot(x='month',y='cases',data=df2group,ax=ax1)
ax1.set_title('Total COVID-19 cases across months in 2020')
ax1.set_xlabel("Months in 2020")
ax1.set_ylabel('Number of cases in million')
ax2=sns.lineplot(x='month',y='deaths',data=df2group,ax=ax2)
ax2.set_title('Total COVID-19 deaths across months in 2020')
ax2.set_xlabel("Months in 2020")
ax2.set_ylabel("Number of deaths")
```

```python
francedf=df1[df1['country']=='France']
```

```python
francedf.reset_index(inplace=True)
```

```python
frdf=francedf.groupby('month')['cases','deaths'].sum()
```

```python
frdf.head()
```

```python
uk=df1[df1['country']=='United_Kingdom']
uk.reset_index(inplace=True)
ukdf=uk.groupby('month')['cases','deaths'].sum()
```

```python
gr=df1[df1['country']=='Germany']
gr.reset_index(inplace=True)
grdf=gr.groupby('month')['cases','deaths'].sum()
```

```python
it=df1[df1['country']=='Italy']
it.reset_index(inplace=True)
itdf=it.groupby('month')['cases','deaths'].sum()
```

```python
fig2,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(20,10))
ax1=sns.lineplot(x='month',y='cases',data=frdf,ax=ax1)
ax1.set_title('Cases in France across months')
ax2=sns.lineplot(x='month',y='cases',data=grdf,ax=ax2)
ax2.set_title('Cases in Germany across months')
ax3=sns.lineplot(x='month',y='cases',data=ukdf,ax=ax3)
ax3.set_title('Cases in United Kingdom across months')
ax4=sns.lineplot(x='month',y='cases',data=itdf,ax=ax4)
ax4.set_title('Cases in Italy across months')
```

```python
fig2,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(20,10))
ax1=sns.lineplot(x='month',y='deaths',data=frdf,ax=ax1)
ax1.set_title('Deaths in France across months')
ax2=sns.lineplot(x='month',y='deaths',data=grdf,ax=ax2)
ax2.set_title('Deaths in Germany across months')
ax3=sns.lineplot(x='month',y='deaths',data=ukdf,ax=ax3)
ax3.set_title('Dearhs in United Kingdom across months')
ax4=sns.lineplot(x='month',y='deaths',data=itdf,ax=ax4)
ax4.set_title('Deaths in Italy across months')
```
